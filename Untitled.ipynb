```python
import pandas as pd

df = pd.read_csv(
    "SMSSpamCollection",
    sep="\t",
    header=None,
    names=["Label", "Message"]
)

df.head(10)
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df[['Label','Message']]
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.duplicated().sum()
```

```python
df['Label']=df['Label'].map({'ham':0,
                             'spam':1
                            })
```

```python
df
```

```python
x = df['Message']
```

```python
y = df['Label']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
x = vectorizer.fit_transform(x)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size = 0.25,
    random_state = 45
)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
g_model= GaussianNB()
m_model = MultinomialNB()
```

```python
b_model = BernoulliNB()
```

```python
g_model.fit(x_train.toarray(),y_train)
```

```python
m_model.fit(x_train.toarray(),y_train)
```

```python
b_model.fit(x_train,y_train)
```

```python
g_pred = g_model.predict(x_test.toarray())
```

```python
m_pred = m_model.predict(x_test)
```

```python
b_pred = b_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('Gaussian_accuracy:',accuracy_score(y_test,g_pred))
print('Multinomial_accuracy:',accuracy_score(y_test,m_pred))
print('Bernoulli_accuracy:',accuracy_score(y_test,b_pred))
```

```python
print('-----Gaussian_report------')
print(classification_report(y_test,g_pred))
print('-----Multinomial_report-----')
print(classification_report(y_test,m_pred))
print('-----Bernoulli_report------')
print(classification_report(y_test,b_pred))
```

```python
print(confusion_matrix(y_test,g_pred))
```

```python
print(confusion_matrix(y_test,m_pred))
```

```python
print(confusion_matrix(y_test,b_pred))
```
